2. How does broadcasting work in NumPy?
  Ans. Broadcasting in numpy refers to the process of apply various arithmetic operation on array of different shapes.

  Broadcasting Rules -

  Numpy compares shape from right to left(trailing dimensions) and apply the following rules-->
  1. if dimension are equal, they are compatible
  2. if one of dimension is 1, then it stretches itself to match other
  3. if both are of diff dimension then error will be thrown.
